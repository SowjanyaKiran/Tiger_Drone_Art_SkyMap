```python
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files
import io
```

```python
# -------------------------------
# Step 1: Upload TXT or CSV file
# -------------------------------
print("Please upload your TXT or CSV file (first line should be height,width,channels):")
uploaded = files.upload()

if not uploaded:
    raise FileNotFoundError("No file uploaded.")
```

```python
# Get the uploaded file name
file_name = list(uploaded.keys())[0]
print(f"Uploaded file: {file_name}")
```

```python
# -------------------------------
# Step 2: Read dimensions and pixel data
# -------------------------------
# Use io.StringIO to read the uploaded file content
file_content = io.StringIO(uploaded[file_name].decode('utf-8-sig'))
```

```python
# Read first line for dimensions
first_line = file_content.readline().strip()
if ',' in first_line:
    height, width, channels = map(int, first_line.split(','))
else:
    height, width, channels = map(int, first_line.split())
```

```python
# Load remaining pixel data
if file_name.endswith(".csv"):
    pixels = np.loadtxt(file_content, delimiter=",", dtype=np.uint8)
else:  # TXT file
    pixels = np.loadtxt(file_content, dtype=np.uint8)
```

```python
# -------------------------------
# Step 3: Reshape to original image
# -------------------------------
regenerated_image = pixels.reshape(height, width, channels)
```

```python
# -------------------------------
# Step 4: Plot the regenerated image
# -------------------------------
plt.figure(figsize=(8, 6))
plt.title("Image Regenerated from Uploaded File")
plt.imshow(regenerated_image)
plt.axis("off")
plt.show()
```
